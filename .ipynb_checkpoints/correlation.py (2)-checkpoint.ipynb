```python
# code to convert the indicators to following format 
'''
    SN      Country        Country Code        Year          % of Agricultural land        Forest Area
    1       Nepal          NEP                 2019           10 
'''

data = {}
rows = []

# print(indicators['Indicator Name'].unique())
    
columns_controls = [
    'Agricultural land (sq. km)',
    'Agricultural irrigated land (% of total agricultural land)',
    'Forest area (sq. km)'
]

columns = ['Country', 'Country Code']
indicator_names = []
for index, row in indicators.iterrows():
    if row['Indicator Name'] not in columns_controls:
        continue
        
    if row['Indicator Name'] not in columns:
        columns.append(row['Indicator Name'])
        
    if row['Indicator Name'] not in indicator_names:
        indicator_names.append(row['Indicator Name'])
        
columns.append('Year')

for index, row in indicators.iterrows():
    if row['Indicator Name'] not in columns_controls:
        continue


    for year in range(1960, 2019):
        data_row = [row['Country Name'], row['Country Code']]
        for each_col in indicator_names:
            data_row.append('')
        
        data_row.append(year)
        rows.append(data_row)

        if row['Indicator Name'] not in data:
            data[row['Indicator Name']] = [row[str(year)]]
        else:
            data [row['Indicator Name']].append(row[str(year)])
    
          
```

```python
# code to convert the indicators to following format 
'''
    Country        Country Code        Year          % of Agricultural land        Forest Area

'''
print(indicator_names)  
```

```python
new_indicators = pandas.DataFrame(rows, columns=columns)


for each_key in data:
    new_indicators[each_key] = pandas.Series(data[each_key])
```

```python
# code to merge temperature and indicators dataframes for candidate indicators after 1991
'''Country       Statistics       Year       Temperature - (Celsius)       Forest area (sq. km)      Agricultural land (sq. km)'''

new_indicators_subset = new_indicators.loc[new_indicators['Year'] >= 1991]
# # df.loc[df['column_name'] == some_value]
# df_forest_area = new_indicators_subset[['Country Code','Forest area (sq. km)', 'Year']]
# df_forest_area.head()

#print(new_indicators_subset.head())

print(new_indicators_subset.loc[new_indicators_subset['Country Code'] == 'PRK'])

column_names = ["Country", "Statistics", "Year", "Temperature - (Celsius)", "Forest area (sq. km)", "Agricultural land (sq. km)"]
df = pandas.DataFrame(columns = column_names)

res = pandas.merge(temperature, new_indicators_subset, how='inner', left_on=['Country Code', ' Year'], right_on = ['Country Code','Year'])

res.head(60)
```
