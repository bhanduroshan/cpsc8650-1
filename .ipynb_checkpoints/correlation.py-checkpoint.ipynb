```python
# code to convert the indicators to following format 
'''
    SN      Country        Country Code        Year          % of Agricultural land        Forest Area
    1       Nepal          NEP                 2019           10 
'''

data = {}
rows = []

columns = ['County', 'Country Code']
indicator_names = []
for index, row in indicators.iterrows():
    if row['Indicator Name'] not in columns:
        columns.append(row['Indicator Name'])
        
    if row['Indicator Name'] not in indicators:
        indicator_names.append(row['Indicator Name'])

columns.append('Year')

for index, row in indicators.iterrows():
    data_row = [row['Country Name'], row['Country Code']]
    for each_col in indicator_names:
        data_row.append('')

    for year in range(1960, 2019):
        data_row.append(year)
        rows.append(data_row)

        if row['Indicator Name'] not in data:
            data[row['Indicator Name']] = [row[str(year)]]
        else:
            data [row['Indicator Name']].append(row[str(year)])
    
          
```

```python
# code to convert the indicators to following format 
'''
    Country        Country Code        Year          % of Agricultural land        Forest Area

'''
print(data.keys())  
```

```python
new_indicators = pandas.DataFrame(rows, columns=columns)

for each_key in data:
    new_indicators[each_key] = new_indicators[each_key].map(data)

new_indicators.head()
```
